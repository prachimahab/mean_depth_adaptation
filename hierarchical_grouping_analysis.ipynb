## Group Stimuli

```python
p1 = '/Users/prachi/Documents/Hierarchical_Grouping_Stimuli/HG_group1'
p2 = '/Users/prachi/Documents/Hierarchical_Grouping_Stimuli/HG_group2'
p3 = '/Users/prachi/Documents/Hierarchical_Grouping_Stimuli/HG_group3'
p4 = '/Users/prachi/Documents/Hierarchical_Grouping_Stimuli/HG_group4'

paths = [p1, p2, p3, p4]

group_stimuli = [[], [], [], []]

for i in range(len(paths)): 
    group_path = paths[i]
    for folder in os.listdir(group_path):
        if folder != '.DS_Store':
            for img in os.listdir(group_path + '/' + folder):
                if img != '.DS_Store':
                    imgpath = group_path + '/' + folder + '/' + img
                    if path.isfile(imgpath) == True:
                        group_stimuli[i].append(img)

for group in group_stimuli:
    print(len(group))
```
